```python
df = pd.get_dummies(df,drop_first=True) 
```

```python
df
```

```python
df.corr()
```

```python

for i in df.columns:
  if((df[i].corr(df["charges"]))<0.1 and (df[i].corr(df["charges"]))> -0.1):
    print(i+" this column can be deleted becuase correlation value is very less and not linearly dependent")
   
```

```python
a=[]
for i in df.columns:
  if((df[i].corr(df["charges"]))>0.1 or (df[i].corr(df["charges"]))< -0.1):
    a.append(i)
```

```python
a
```

```python
df_new=df[a]
```

```python
df_new
```

```python
a=["charges"]
b=["age","bmi","smoker_yes"]
X=df_new[b]
y=df_new[a]
```

```python
from sklearn.model_selection import train_test_split
(X_train,X_test,y_train,y_test) = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
clf.fit(X_train,y_train)
```

```python
c=clf.coef_
c
```

```python
i=clf.intercept_
i
```

```python
#charges="age"*264.244+"bmi"*332.387+"smoker_yes"*23660.54-12059
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
print('MSE of Linear Regression : ' + str(round(mean_squared_error(y_test,y_pred),2)))
```
