```python
df['CHAS'] = df['CHAS'].astype('category')
df['RAD'] = df['RAD'].astype('category')
```

```python
df.info()
```

```python
np.shape(df)
```

```python
y = df['MEDV']
X = df.loc[:, df.columns != 'MEDV']

print(np.shape(y))
print(np.shape(X))
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn import preprocessing

ss = preprocessing.StandardScaler()
X_train =  ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
print(f'X_train: {np.shape(X_train)}\nY_train: {np.shape(y_train)}\nX_test: {np.shape(X_test)}\ny_test: {np.shape(y_test)}')
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
print('MSE of Linear Regression : ' + str(round(mean_squared_error(y_test,y_pred),2)))
```

```python

df_preds = pd.DataFrame({'Truth: ': y_test, 'Predicted: ': y_pred})
df_preds.head(10)
```

```python
import matplotlib.pyplot as plt  

plt.scatter(y_test, y_pred, label='Predictions')
plt.plot(y_test, y_test, color='red', linewidth=2, label='regression line')
plt.title('Linear regression')
plt.xlabel('y_test')
plt.ylabel('y_Predictions')
plt.legend(loc="upper left")
```
