 kNN methods is to identify 'k' samples in the dataset that are similar or close in the space. We use these 'k' samples to estimate the value of the missing data points. Each sample's missing values are imputed using the mean value of the 'k'-neighbors found in the dataset

```python
from sklearn.impute import SimpleImputer
imputer_int = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer_str = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
df.columns
```

```python
df.info()
```

```python
numcols=['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation',
       'Sunshine','WindGustSpeed',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm']
```

```python
catcols=["WindGustDir","WindDir9am","WindDir3pm","RainToday","RainTomorrow"]
```

```python
for i in numcols:
  df[i] = imputer_int.fit_transform(df[[i]])
```

```python
for j in catcols:
  df[j] = imputer_str.fit_transform(df[[j]])
```

```python
df.drop(['Date', 'Location'], inplace = True,axis = 1)
```

```python
from sklearn.preprocessing import LabelBinarizer
LB = LabelBinarizer()
df['WindGustDir'] = LB.fit_transform(df[['WindGustDir']])
df['WindDir9am'] = LB.fit_transform(df[['WindDir9am']])
df['WindDir3pm'] = LB.fit_transform(df[['WindDir3pm']])
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df['RainToday'] = LE.fit_transform(df['RainToday'])
df['RainTomorrow'] = LE.fit_transform(df['RainTomorrow'])
```

```python
df.head(10)
```

```python
X = df.drop('RainTomorrow',axis=1).values
y = df['RainTomorrow'].values
```

```python
from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
print('Shape of X training set {}'.format(X_trainset.shape),'&',' Size of Y training set {}'.format(y_trainset.shape))
```

desicion tree

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion="entropy")
clf
```

```python
clf.fit(X_trainset,y_trainset)
```

```python
pred = clf.predict(X_testset)
pred[0:5]
```

```python
print(pred[0:5])
print(y_testset[0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
print("classificationreport")
print(classification_report(y_testset,pred,digits=2))
print("confusion marix")
print(confusion_matrix(y_testset,pred))
from sklearn.metrics import jaccard_score
print("jaccob index")
print(jaccard_score(y_testset,pred, average='weighted'))
```

knn

```python
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier(n_neighbors=5)
clf.fit(X_trainset, y_trainset)
y_pred = clf.predict(X_testset)
print("\n Predicted Values:\n")
print(y_pred [0:5])
from sklearn.metrics import classification_report, confusion_matrix
import itertools
print("classificationreport")
print(classification_report(y_testset,pred,digits=2))
print("confusion marix")
print(confusion_matrix(y_testset,pred))
from sklearn.metrics import jaccard_score
print("jaccob index")
print(jaccard_score(y_testset,pred, average='weighted'))
```

```python
from sklearn import metrics

print("knn Accuracy: ", metrics.accuracy_score(y_testset,y_pred))
```
